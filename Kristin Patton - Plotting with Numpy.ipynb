```python
import numpy as np
import matplotlib.pyplot as plt

sig = 2
u0 = 0.002
er = 1e-5
v0 = u0 + er

dt = 0.001
t = np.arange(0, 10 + dt, dt)

U = (u0*np.exp(sig*t))/(1 - u0 + u0*np.exp(sig*t))
V = (v0*np.exp(sig*t))/(1 - v0 + v0*np.exp(sig*t))

#(a1)
a1 = np.log(U[1:]/U[:-1])/dt

#(a2)
a2 = sig*(1 - U)

#(b1)
b1 = np.log((U[1:] - V[1:])/(U[:-1] - V[:-1]))/dt

#(b2)
b2 = sig*(1 - 2*U)

plt.subplot(1,4,1)
plt.plot(t, U)

plt.subplot(1,4,2)
plt.plot(t, U-V)

plt.subplot(1,4,3)
plt.plot(t, a2, 'r.')
plt.plot(t[:-1], a1, 'b-')

plt.subplot(1,4,4)
plt.plot(t, b2, 'r.')
plt.plot(t[:-1], b1, 'b-')

plt.tight_layout()
plt.show()
```
